# Co- Varince 


if covarince of two list or array is positive then both are incresing in order according to each other.
if covarince of two list or array is negative then both are dissimilar and one is incresing then other is dicresing in order.

```python
import matplotlib.pyplot as plt
```

```python
def cov(x,y):
    l=[]
    for i in range(len(x)):
        num=(x[i]-(sum(x)/len(x)))*(y[i]-(sum(y)/len(y)))
        l.append(num)
    return sum(l)
x=[1,2,3,4,5,6]
y=[7,8,9,10,11,12]
print(cov(x,y))

plt.plot(x,y)
plt.show()
```
